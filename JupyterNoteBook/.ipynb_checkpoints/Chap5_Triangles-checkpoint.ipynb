### üß™ Atelier 2 ‚Äì M√©thode des triangles (Interpolation par moyenne locale)

D√©placez le point rouge dans l‚Äôespace 2D pour voir comment la m√©thode des triangles estime une valeur.

- La **valeur estim√©e** est la **moyenne simple** des trois valeurs situ√©es aux **sommets du triangle contenant le point**.
- Ce triangle provient d'une **triangulation de Delaunay** sur les points d'√©chantillonnage.
- Observez comment l‚Äôestimation varie selon la position et les valeurs locales.

> üîÅ Essayez des points proches des bords ou entre des zones tr√®s contrast√©es.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown

triangle_pts = np.array([
    [0, 0],  # A
    [50, 0],  # B
    [25, 25]   # C
])

z_vals = np.array([5.1, 3.7, 4.1])

def barycentric_coords(p, tri):
    A = tri[0]
    B = tri[1]
    C = tri[2]

    v0 = B - A
    v1 = C - A
    v2 = p - A

    denom = v0[0]*v1[1] - v1[0]*v0[1]
    if abs(denom) < 1e-12:
        return np.array([-1, -1, -1])  # Triangle d√©g√©n√©r√©

    w2 = (v2[0]*v1[1] - v1[0]*v2[1]) / denom
    w3 = (v0[0]*v2[1] - v2[0]*v0[1]) / denom
    w1 = 1 - w2 - w3
    return np.array([w1, w2, w3])

def interpolate_value(p, tri, vals):
    w = barycentric_coords(p, tri)
    return np.dot(w, vals), w

def plot_interpolation_with_equations(px, py):
    p = np.array([px, py])
    val_interp, w = interpolate_value(p, triangle_pts, z_vals)
    clear_output(wait=True)

    fig, ax = plt.subplots(figsize=(7,7))
    tri_closed = np.vstack([triangle_pts, triangle_pts[0]])
    ax.plot(tri_closed[:,0], tri_closed[:,1], 'b-', linewidth=2, label='Triangle')

    for i, (x, y) in enumerate(triangle_pts):
        ax.plot(x, y, 'ko', markersize=10)
        ax.text(x+1, y+1, f"Sommet {i}\nZ={z_vals[i]:.1f}", color='black', fontsize=12)

    ax.plot(px, py, 'rx', markersize=12, label=f'Point √† estimer\nZ={val_interp:.2f}')
    ax.text(px+1, py+1, f"Z interpol√© = {val_interp:.2f}", fontsize=12, color='red')

    ax.set_xlim(-10, 60)
    ax.set_ylim(-10, 40)
    ax.set_aspect('equal')
    ax.set_title("Interpolation lin√©aire dans un triangle", fontsize=14)
    ax.legend()
    plt.grid(True)
    plt.show()


x_slider = widgets.FloatSlider(value=25, min=10, max=80, step=0.5, description='X coord:')
y_slider = widgets.FloatSlider(value=5, min=5, max=80, step=0.5, description='Y coord:')

def on_change(x, y):
    w = barycentric_coords(np.array([x, y]), triangle_pts)
    if np.all(w >= -1e-6) and np.all(w <= 1 + 1e-6):
        plot_interpolation_with_equations(x, y)
    else:
        clear_output(wait=True)
        print("Le point est hors du triangle, d√©placez-le √† l'int√©rieur.")

out = widgets.interactive_output(on_change, {'x': x_slider, 'y': y_slider})

display(x_slider, y_slider, out)

plot_interpolation_with_equations(x_slider.value, y_slider.value)


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay

def barycentric_coords(p, tri_pts):
    A = tri_pts[0]
    B = tri_pts[1]
    C = tri_pts[2]
    v0 = B - A
    v1 = C - A
    v2 = p - A
    denom = v0[0]*v1[1] - v1[0]*v0[1]
    if abs(denom) < 1e-12:
        return np.array([-1, -1, -1])
    w2 = (v2[0]*v1[1] - v1[0]*v2[1]) / denom
    w3 = (v0[0]*v2[1] - v2[0]*v0[1]) / denom
    w1 = 1 - w2 - w3
    return np.array([w1, w2, w3])

np.random.seed(42)
n_points = 8
x_samples = np.random.uniform(10, 90, n_points)
y_samples = np.random.uniform(10, 90, n_points)
z_samples = np.round(np.random.uniform(0, 10, n_points), 2)
points = np.vstack((x_samples, y_samples)).T

delaunay = Delaunay(points)

resolution = 200
grid_x, grid_y = np.meshgrid(np.linspace(0, 100, resolution), np.linspace(0, 100, resolution))
grid_points = np.c_[grid_x.ravel(), grid_y.ravel()]

simplex_indices = delaunay.find_simplex(grid_points)
est_values = np.full(grid_points.shape[0], np.nan)

for i, s in enumerate(simplex_indices):
    if s != -1:
        tri_idx = delaunay.simplices[s]
        tri_pts = points[tri_idx]
        vals = z_samples[tri_idx]
        p = grid_points[i]
        w = barycentric_coords(p, tri_pts)
        if np.all(w >= -1e-6) and np.all(w <= 1+1e-6):
            est_values[i] = np.dot(w, vals)

est_grid = est_values.reshape((resolution, resolution))

plt.figure(figsize=(10, 8))

# Affichage carte interpol√©e
contour = plt.contourf(grid_x, grid_y, est_grid, levels=100, cmap='viridis', alpha=0.85)

# Affichage triangulation
plt.triplot(points[:,0], points[:,1], delaunay.simplices, color='white', linewidth=1.2, alpha=0.8)

# Points d'√©chantillonnage avec annotations valeurs
sc = plt.scatter(x_samples, y_samples, c=z_samples, s=120, cmap='viridis', edgecolor='black', zorder=5)
for i, (x,y,z) in enumerate(zip(x_samples, y_samples, z_samples)):
    plt.text(x+1, y+1, f"{z:.1f}", color='white', fontsize=12, weight='bold', zorder=6)

plt.colorbar(sc, label='Valeurs mesur√©es')
plt.title("Interpolation lin√©aire dans triangles - M√©thode des triangles", fontsize=16, weight='bold')
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.grid(alpha=0.3)
plt.gca().set_aspect('equal')

plt.show()
```
