```python
# Ateliers pédagogiques : Méthodes usuelles d’estimation spatiale (1D ou 2D)

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay
import ipywidgets as widgets

np.random.seed(42)

# -----------------------------
# 1. Données de base 2D
# -----------------------------
n_points = 8
x_samples = np.random.uniform(10, 90, n_points)
y_samples = np.random.uniform(10, 90, n_points)
z_samples = np.round(np.random.uniform(0, 10, n_points), 2)

points = np.vstack((x_samples, y_samples)).T

# -----------------------------
# 2. Atelier 1 : Méthode des polygones de Voronoï
# -----------------------------
def plot_polygone():
    vor = Voronoi(points)
    fig, ax = plt.subplots(figsize=(6,6))
    voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='gray', line_width=1, line_alpha=0.6, point_size=2)
    sc = ax.scatter(x_samples, y_samples, c=z_samples, s=200, cmap='viridis', edgecolor='k')
    for i, txt in enumerate(z_samples):
        ax.annotate(f"{txt}", (x_samples[i]+1, y_samples[i]+1))
    plt.colorbar(sc, label="Valeur attribuée")
    ax.set_title("Méthode des polygones (Voronoï)")
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.set_aspect('equal')
    plt.grid(True)
    plt.show()

# -----------------------------
# 3. Atelier 2 : Méthode des triangles (moyenne locale dans triangle)
# -----------------------------
def plot_triangle(x0=50, y0=50):
    delaunay = Delaunay(points)
    simplex = delaunay.find_simplex([[x0, y0]])
    triangle = delaunay.simplices[simplex][0]

    tri_pts = points[triangle]
    tri_vals = z_samples[triangle]
    val_est = np.mean(tri_vals)

    fig, ax = plt.subplots(figsize=(6,6))
    ax.triplot(x_samples, y_samples, delaunay.simplices, color='lightgray')
    ax.scatter(x_samples, y_samples, c=z_samples, s=200, cmap='viridis', edgecolor='k')
    ax.scatter(x0, y0, marker='x', color='red', s=100, label='Point estimé')
    ax.fill(tri_pts[:,0], tri_pts[:,1], edgecolor='blue', fill=False, linewidth=2)
    ax.annotate(f"Estimation = {val_est:.2f}", (x0+2, y0+2), fontsize=12)
    ax.set_title("Méthode des triangles")
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.legend()
    ax.set_aspect('equal')
    plt.grid(True)
    plt.show()

# -----------------------------
# 4. Atelier 3 : Inverse de la distance pondérée (IDW)
# -----------------------------
def plot_idw(x0=50, y0=50, power=2.0):
    dists = np.sqrt((x_samples - x0)**2 + (y_samples - y0)**2)
    with np.errstate(divide='ignore'):  # éviter division par 0
        weights = 1 / dists**power
    weights[dists == 0] = 1e10  # poids élevé si pile sur un point
    weights /= np.sum(weights)
    val_est = np.sum(weights * z_samples)

    fig, ax = plt.subplots(figsize=(6,6))
    ax.scatter(x_samples, y_samples, c=z_samples, s=200, cmap='viridis', edgecolor='k')
    ax.scatter(x0, y0, marker='x', color='red', s=100, label='Point estimé')
    for i, w in enumerate(weights):
        ax.plot([x_samples[i], x0], [y_samples[i], y0], linestyle='--', color='gray', alpha=0.5)
    ax.annotate(f"Estimation = {val_est:.2f}", (x0+2, y0+2), fontsize=12)
    ax.set_title(f"Méthode Inverse Distance (p={power})")
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.legend()
    ax.set_aspect('equal')
    plt.grid(True)
    plt.show()

# -----------------------------
# Widgets interactifs pour les ateliers 2 et 3
# -----------------------------
x_slider = widgets.FloatSlider(value=50, min=0, max=100, step=1, description="X")
y_slider = widgets.FloatSlider(value=50, min=0, max=100, step=1, description="Y")
power_slider = widgets.FloatSlider(value=2.0, min=0.5, max=5.0, step=0.1, description="Puissance")

# Lancer les ateliers
print("\n▶ Atelier 1 : Méthode des polygones de Voronoï")
plot_polygone()

print("\n▶ Atelier 2 : Méthode des triangles (interactive)")
widgets.interact(plot_triangle, x0=x_slider, y0=y_slider);

print("\n▶ Atelier 3 : Méthode de l’inverse de la distance (interactive)")
widgets.interact(plot_idw, x0=x_slider, y0=y_slider, power=power_slider);
```
